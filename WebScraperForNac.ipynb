```python
def saver(nam):
    url="https://wms.lroc.asu.edu/lroc/view_lroc/LRO-L-LROC-3-CDR-V1.0/"+str(nam)
    content=requests.get(url)
    htmlcontent=content.content

    #print(soup.prettify)
    lis=["Upper left latitude","Upper left longitude","Upper right latitude","Upper right longitude","Lower left latitude","Lower left longitude","Lower right latitude","Lower right longitude"]
    val=[nam]
    temp_link="https:"+str(htmlcontent).split("\">Download EDR")[0].split("<li><a href=\"")[-1]
    print(temp_link)
    
    for i in lis:
        s=str(htmlcontent).split(i+"</th>\\n        <td>")
        s2=s[1].split("</td>")
        val.append(round(float(s2[0]),2))
    smaller_sq=val
    final.loc[len(final.index)]=val
    #print(val)
    
    
```

```python
import threading
```

```python
file=open("Reader2.txt",'r')
useful_links=file.readlines()
```

```python
up=10

threads=[]
for t in range(9):
    for i in range(1+t*up,up+t*up):
        s=useful_links[i][:-1].split('/')[-1]
        s=s[:-5]+'C'+'.IMG'
        t=threading.Thread(target=saver,args=(s,))
        threads.append(t)
        t.start()
for t in threads:
    t.join()
    print('')
    val=7
```
